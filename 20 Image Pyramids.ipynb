##### Image Pyramids

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('images/sachin_new.jpg')
pyramids = [img]
for i in range(5):
    img = cv2.pyrDown(img)
    pyramids.append(img)
len(pyramids)
```

```python
plt.figure(figsize=(18, 9))
for index, pyramid in enumerate(pyramids):
    plt.subplot(1, len(pyramids), index + 1)
    plt.title(f'Image - {index + 1}')
    plt.imshow(pyramid)
    plt.imsave(f'sachin{index}.jpg', pyramid)
```

```python
for i in range(6):
    plt.imsave('sachin%d.jpg' %i, pyramids[i])
```
