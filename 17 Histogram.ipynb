##### Histogram

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

##### Simple Histogram

```python
marks = np.array([56, 76, 89, 87, 65, 61, 65, 45, 69, 71, 72, 42, 55, 67, 81, 69, 68, 61])
plt.title('Histogram')
plt.grid()
plt.xlabel('Marks')
plt.xticks(range(0, 100, 10))
plt.hist(marks, bins=range(0, 100, 10), color='r')
```

##### Image Histogram

```python
img_grey = cv2.imread('images/sachin.jpg', 0)
plt.title('Histogram')
plt.grid()
plt.xlabel('Pixels')
plt.hist(img_grey.flatten(), bins=range(0, 256), color='b'); 
```

```python
img = cv2.imread('images/sachin.jpg')
B, G, R = cv2.split(img)
B.shape, G.shape, R.shape
```

```python
plt.title('RGB Histogram')
plt.grid()
plt.xlabel('Pixels')
plt.hist(B.flatten(), bins=range(0, 256), color='b')
plt.hist(G.flatten(), bins=range(0, 256), color='g')
plt.hist(R.flatten(), bins=range(0, 256), color='r'); 
```
