#### 2. Use the given file named cricket.csv and perform the following operations:

##### 1. Read the file in DataFrame

```python
df = pd.read_csv('datasets/cricket.csv')
df
```

##### 2. List the name of cricketer and their respective runs

```python
df[['name', 'runs']]
```

##### 3. Find total wickets taken by them

```python
df['wickets'].sum()
```

##### 4. Find average of catches taken

```python
df['catches'].mean()
```

##### 5. Calculate strike rates of all the batsmen

```python
# Assuming 100 balls faced per match (as balls faced is not given)
df['ball_faced'] = df['matches'] * 100

# Calculate strike rate and display results
df['strike_rate'] = ((df['runs'] / df['ball_faced']) * 100).round(2)
df
```

##### 6. Print the names of person who has strike rate greater than 50

```python
df[df['strike_rate'] > 50]
```

##### 7. Find the name of wicketkeeper

```python
df[df['stumpings'] != 0]['name']
```

##### 8. Print the name of bowler who played highest number of matches

```python
df[df['wickets'] != 0].sort_values('matches', ascending=False).loc[0, 'name']
```

##### 9. Find average of all the bowlers

```python
df[df['wickets'] != 0]['wickets'].mean()
```

##### 10. Find name of the bowler with least bowling average

```python
df['bowling_average'] = df['runs'] / df['wickets']
df.sort_values('bowling_average')['name'].values[0]
```

##### 11. Print information about all players whose catches per match ratio is greater than 0.5 store it in a csv file

```python
df['catches_per_match'] = df['catches'] / df['matches']
df[df['catches_per_match'] > 0.5].to_csv('datasets/output.csv', index=False)
```
