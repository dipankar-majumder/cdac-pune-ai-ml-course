##### Pixel Operations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros([500, 500], dtype=np.uint8)
plt.imshow(img, cmap='gray')
```

```python
img = np.zeros([500, 500], dtype=np.uint8)
img[100, 100] = 0xFF
plt.imshow(img, cmap='gray')
```

```python
img = np.zeros([500, 500], dtype=np.uint8)
img[100, :] = 0xFF
plt.imshow(img, cmap='gray')
```

```python
img = np.zeros([500, 500], dtype=np.uint8)
img[:, 100] = 0xFF
plt.imshow(img, cmap='gray')
```

```python
img = np.zeros([500, 500], dtype=np.uint8)
map(lambda i: lambda j: 0xFF if i == j else i, img)
plt.imshow(img, cmap='gray')
img
```

```python
for i in range(500):
    img[i, i] = 0xFF
plt.imshow(img, cmap='gray')
```
