##### Image Flipping

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg')
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
help(cv2.flip)
```

```python
img_ver = cv2.flip(img, 0)
img_hor = cv2.flip(img, 1)
img_dia = cv2.flip(img, -1)
```

```python
cv2.imshow('Horizontal', img_hor)
cv2.imshow('Vertical', img_ver)
cv2.imshow('Diagonal', img_dia)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ver_flip = cv2.cvtColor(img_ver, cv2.COLOR_BGR2RGB)
hor_flip = cv2.cvtColor(img_hor, cv2.COLOR_BGR2RGB)
dia_flip = cv2.cvtColor(img_dia, cv2.COLOR_BGR2RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(16, 9))
plt.subplot(1, 4, 1); plt.title('Original'); plt.imshow(img_rgb)
plt.subplot(1, 4, 2); plt.title('Original'); plt.imshow(ver_flip)
plt.subplot(1, 4, 3); plt.title('Original'); plt.imshow(hor_flip)
plt.subplot(1, 4, 4); plt.title('Original'); plt.imshow(dia_flip)
```
