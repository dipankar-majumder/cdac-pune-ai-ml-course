##### Data Visualisation using Matplotlib

##### import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### Line Plot

```python
# Define the data
x = np.array([0, 11, 23, 26, 43, 48, 56, 61, 68, 78])
y = np.array([10, 14, 25, 34, 46, 51, 50, 68, 78, 81])
z = np.array([67, 65, 59, 47, 43, 38, 35, 32, 19, 13])
p = np.array([56, 43, 47, 65, 12, 47, 65, 45, 66, 75])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.plot(y, z)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xlim(0, 90)
plt.ylim(0, 90)
plt.xticks(range(0, 90, 5))
plt.yticks(range(0, 90, 5))
plt.plot(x, y)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xlim(0, 90)
plt.ylim(0, 90)
plt.xticks(range(0, 90, 5))
plt.yticks(range(0, 90, 5))
# markers: '.', 'o', '*', '+', '<', '>', 'v', '^', 'D'
# ls: '-', '--', '-.'
plt.plot(x, y, color='red', marker='*', ls='-.', ms=10, mfc='green', lw=2)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, color='r')
plt.plot(x, z, color='g')
```

```python
plt.title('My first plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, color='r', label=2023)
plt.plot(x, z, color='g', label=2024)
plt.legend(loc='upper right', bbox_to_anchor=(0.5, 0.0, 0.5, 0.5))
```

```python
plt.title('Comparison plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, color='r', label=2023)
for i in range(10):
    plt.annotate(text=f'({x[i]}, {y[i]})', xy=(x[i] - 2, y[i] + 2))
```

```python
plt.figure(figsize=(7, 7))
plt.subplot(2, 2, 1);   plt.title('First');     plt.plot(x, y, color='r')
plt.subplot(2, 2, 2);   plt.title('Second');    plt.plot(x, z, color='g')
plt.subplot(2, 2, 3);   plt.title('Third');     plt.plot(x, p, color='b')
plt.subplot(2, 2, 4);   plt.title('Fourth');    plt.plot(y, z, color='y')
```

```python
plt.figure(figsize=(7, 7))
plt.subplot(2, 2, 1);       plt.title('First');     plt.plot(x, y, color='r')
plt.subplot(2, 2, (2, 4));  plt.title('Second');    plt.plot(x, z, color='g')
plt.subplot(2, 2, 3);       plt.title('Third');     plt.plot(x, p, color='b')
```

```python
plt.figure(figsize=(7, 7))
plt.subplot(2, 2, 1);   plt.title('First');     plt.plot(x, y, color='r')
plt.subplot(1, 2, 2);   plt.title('Second');    plt.plot(x, z, color='g')
plt.subplot(2, 2, 3);   plt.title('Third');     plt.plot(x, p, color='b')
```

```python
student = pd.read_csv('datasets/student.csv')
```

```python
plt.plot(student['roll'], student['marks'], marker='.', ms=4)
for student_roll, student_name, student_marks in student.loc[:, ('roll', 'name', 'marks')].values:
    plt.annotate(text=student_name, xy=(student_roll, student_marks))
```

```python
plt.plot(student['roll'], student['marks'], marker='o')
for i in range(10):
    plt.annotate(text=student['name'].values[i], xy=(student['roll'].values[i], student['marks'].values[i]))
```

##### Bar Plot

```python
studs = [61, 79, 156, 189, 104, 111, 58]
langs = ['C', 'C++', 'Java', 'Python', 'PHP', 'JavaScript', 'R']
```

```python
plt.title('Student Enrollment')
plt.xlabel('Course')
plt.ylabel('Total Students')
plt.bar(langs, studs, color='m')
```

##### Horizontal Bar

```python
plt.title('Student Enrollment')
plt.xlabel('Total Students')
plt.ylabel('Course')
plt.barh(langs, studs, color='m')
```

```python
plt.title('Student Enrollment')
plt.xlabel('Marks')
plt.ylabel('Student Name')
plt.barh(student['name'], student['marks'], color='navy')
```

```python
plt.title('Student Enrollment')
plt.xlabel('Marks')
plt.ylabel('Student Name')
plt.bar(student['name'], student['marks'] + student['age'], color='yellow', label='age')
plt.bar(student['name'], student['marks'], color='blue', label='marks')
plt.legend()
```

```python
plt.title('Student Enrollment')
plt.xlabel('Marks')
plt.ylabel('Student Name')
plt.bar(student['name'], student['marks'], color='orange', label='marks', align='center')
plt.bar(student['name'], student['age'], color='cyan', label='age', align='edge', width=0.5)
plt.legend()
```

##### Pie Chart
