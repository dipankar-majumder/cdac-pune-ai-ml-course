##### Naive Bayes Classification

##### Dataset

Download `divorce.csv`

Location: https://mitu.co.in/dataset

Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('datasets/divorce.csv', sep=';')
df
```

```python
df.shape
```

Separate the input and output data

```python
# input data
x = df.drop('Class', axis=1)

# output data
y = df['Class']
```

```python
x.describe()
```

```python
sns.countplot(y=y)
```

```python
plt.pie(y.value_counts(), labels=['Not Divorsed', 'Divorsed'], autopct='%.2f%%'); 
```

##### Cross-Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

##### Build the model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(x_train, y_train)
```

##### Performance Evaluation

```python
y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
nb.predict_proba(x_test)
```

```python
nb.predict(new)
```

##### AUC - ROC Curve

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
y_prob = nb.predict_proba(x_test)[::, 1]
y_prob
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob)
```

```python
fpr, tpr, _
```

```python
auc = roc_auc_score(y_test, y_prob)
auc
```

```python
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.grid()
plt.plot(fpr, tpr)
```

##### Comparing with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)

y_log = log.predict(x_test)
y_log_prob = log.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_log_prob)
auc = roc_auc_score(y_test, y_log_prob)
print('AUC:', auc)
print('Accuracy:', accuracy_score(y_test, y_log))
```

```python
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.grid()
plt.plot(fpr, tpr)
```
