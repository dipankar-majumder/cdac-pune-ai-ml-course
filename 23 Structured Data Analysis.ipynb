##### Join, Merge, Append, Concat

```python
df = pd.read_csv('datasets/student.csv')
```

```python
newdf = pd.DataFrame({
    'roll': range(1, 11),
    'cgpa': df['marks'] / 10 + 0.47,
    'gender': ['M', 'M', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'M']
})
newdf
```

```python
pd.merge(df, newdf)
```

```python
df.columns
```

```python
newdf.columns
```

```python
df.merge(newdf, on='roll')
```

```python
df5 = pd.read_csv('datasets/student5.csv')
df5
```

```python
cdf = pd.concat([df, df5])
cdf
```
