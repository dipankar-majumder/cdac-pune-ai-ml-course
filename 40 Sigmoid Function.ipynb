##### Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(-5, 5, 0.25)
x
```

```python
plt.plot(x)
```

```python
def sigmoid(y):
    z = 1 / (1 + np.exp(-y))
    return z
```

```python
plt.plot(x, sigmoid(x))
```

```python
x = np.arange(-10, 10, 0.25)
x
```

```python
plt.plot(x, sigmoid(x))
```

```python
sigmoid(np.inf), sigmoid(-np.inf)
```
