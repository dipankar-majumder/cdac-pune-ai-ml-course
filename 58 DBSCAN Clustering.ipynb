##### DBSCAN Clustering

Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
help(make_blobs)
```

```python
x, y = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.6, random_state=0)
x
```

```python
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.scatter(x[:, 0], x[:, 1], c=y)
```

##### Build the DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
help(DBSCAN)
```

```python
dbs = DBSCAN(eps=0.6, min_samples=4)
```

```python
labels = dbs.fit_predict(x)
labels
```

##### Visualise the clusters

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title('Unclustered Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.scatter(x[:, 0], x[:, 1], c=y)

plt.subplot(1, 2, 2)
plt.title('Unclustered Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.scatter(x[:, 0], x[:, 1], c=labels)
```

```python
df = pd.DataFrame(data=x, columns=['Feature 1', 'Feature 2'])
df['class'] = labels
```

```python
sns.jointplot(data=df, x='Feature 1', y='Feature 2', hue='class', palette='tab10')
```

```python
anomalies = df[df['class'] == -1]
anomalies
```
