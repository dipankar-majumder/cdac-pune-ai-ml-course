##### Data Encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/student.csv')
df
```

##### Label Encoder

```python
# import the class
from sklearn.preprocessing import LabelEncoder
```

```python
# create the object
encoder = LabelEncoder()
```

```python
# apply the method
df['name'] = encoder.fit_transform(df['name'])
df
```

```python
encoder.classes_
```

```python
encoder.transform(['amol'])
```

```python
encoder.inverse_transform(df['name'])
```
