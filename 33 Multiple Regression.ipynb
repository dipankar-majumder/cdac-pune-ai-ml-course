##### Dataset

Download `mtcars.csv`

Location: https://mitu.co.in/dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Load the data

```python
df = pd.read_csv('datasets/mtcars.csv')
df
```

Separate the input and output variables

```python
# input data
x = df[['disp', 'hp', 'wt']]

# output data
y = df['mpg']
```

```python
x
```

```python
import seaborn as sns
sns.kdeplot(y)
```

Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
x.corrwith(y)
```

```python
disp, hp, wt = 221, 102, 200.91
disp * regressor.coef_[0] + hp * regressor.coef_[1] + wt * regressor.coef_[2] + regressor.intercept_
```

```python
regressor.predict([[disp, hp, wt]])
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.title('disp vs mpg')
plt.xlabel('mpg')
plt.ylabel('disp')
plt.scatter(x['disp'], y)
plt.subplot(2, 2, 2)
plt.title('hp vs mpg')
plt.xlabel('mpg')
plt.ylabel('hp')
plt.scatter(x['hp'], y)
plt.subplot(2, 2, 3)
plt.title('wt vs mpg')
plt.xlabel('mpg')
plt.ylabel('wt')
plt.scatter(x['wt'], y)
```
