```python
##### Thresholding
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
new_img = img > 127
inv_img = img <= 127
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1);   plt.title('Original Image');                plt.imshow(img, cmap='Greys_r')
plt.subplot(1, 3, 2);   plt.title('Thresholded Image');             plt.imshow(new_img, cmap='Greys_r')
plt.subplot(1, 3, 3);   plt.title('Inverted Thresholded Image');    plt.imshow(inv_img, cmap='Greys_r')
```

```python
##### Otsu Thresholding
```

```python
help(cv2.threshold)
```

```python
otsu_thresh, otsu_img = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
otsu_thresh
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1, 4, 1);   plt.title('Original Image');                plt.imshow(img, cmap='Greys_r')
plt.subplot(1, 4, 2);   plt.title('Thresholded Image');             plt.imshow(new_img, cmap='Greys_r')
plt.subplot(1, 4, 3);   plt.title('Inverse Thresholded Image');     plt.imshow(inv_img, cmap='Greys_r')
plt.subplot(1, 4, 4);   plt.title('Otsu Thresholded Image');        plt.imshow(otsu_img, cmap='Greys_r')
```
