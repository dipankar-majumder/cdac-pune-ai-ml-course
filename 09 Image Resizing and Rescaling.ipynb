##### Image Resizing and Rescaling

##### Using OpenCV

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg')
```

```python
help(cv2.resize)
```

```python
img_new = cv2.resize(img, (300, 500))
img_new.shape
```

```python
cv2.imshow('Resized Image', img_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_new = cv2.resize(img, None, fx=1.2, fy=0.75)
```

```python
cv2.imshow('Resized Image', img_new)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_new = cv2.resize(img, None, fx=1.2, fy=0.75)
```

##### Using skimage

```python
import skimage.transform
```

```python
img_rgb = plt.imread('images/sachin.jpg')
img_new = skimage.transform.resize(img_rgb, (450, 300))
```

```python
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)
plt.subplot(1, 2, 2)
plt.title('Resized Image')
plt.imshow(img_new)
plt.savefig('Resized.png')
```
