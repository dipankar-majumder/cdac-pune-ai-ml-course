##### Load the video

```python
vid = cv2.VideoCapture('images/nature.mp4')
```

```python
height = vid.get(cv2.CAP_PROP_FRAME_HEIGHT)
weight = vid.get(cv2.CAP_PROP_FRAME_WIDTH)
count = vid.get(cv2.CAP_PROP_FRAME_COUNT)
fps = vid.get(cv2.CAP_PROP_FPS)
vid_type = vid.get(cv2.CAP_PROP_FRAME_TYPE)

height, weight, count, fps, vid_type
```

```python
vid = cv2.VideoCapture('images/nature.mp4')
while True:
    ret, img = vid.read()
    if not ret:
        break
    cv2.imshow('Video', img)
    if cv2.waitKey(int(1000 / vid.get(cv2.CAP_PROP_FPS))) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
timestamp = '00:00:40'
hrs, mins, secs = [int(i) for i in timestamp.split(':')]
hrs, mins, secs
```

```python
trigger_frame = (hrs * 60 ** 2 + mins * 60 ** 1 + secs * 60 ** 0) * fps
trigger_frame
```

```python
vid.set(1, trigger_frame)
```

```python
ret, img = vid.read()
```

```python
cv2.imshow('Frame', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
