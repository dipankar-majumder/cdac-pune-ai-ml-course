##### Image Reconstruction Inplainting

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/cat_damaged.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
height, width, depth = img.shape
height, width, depth
```

```python
mask = np.zeros((height, width, depth), dtype=np.uint8)
mask = np.bitwise_not(mask)
plt.imshow(mask)
```

```python
new_img = cv2.bitwise_xor(img_rgb, mask)
plt.imshow(new_img)
```

```python
mask = np.zeros((height, width, depth), dtype=np.uint8)

for i in range(width):
    for j in range(height):
        if sum(img[i, j]) == 0:
            mask[i, j] = (255, 255, 255)
plt.imshow(mask)
```

```python
help(cv2.inpaint)
```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2GRAY)
```

```python
# Apply the inpainting
img_new = cv2.inpaint(img_rgb, mask, 3, cv2.INPAINT_NS)
plt.subplot(1, 2, 1); plt.title('Original Image');      plt.imshow(img_rgb)
plt.subplot(1, 2, 2); plt.title('Reconstructed Image'); plt.imshow(img_new); 
```
