##### Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('datasets/Salary_Data.csv')
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x, y)
print('Accuracy:', regressor.score(x, y))
```

##### Polynomial Transformation

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
x_poly
```

```python
regressor = LinearRegression()
regressor.fit(x_poly, y)
print('Accuracy:', regressor.score(x_poly, y))
```

```python
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)
```

```python
regressor = LinearRegression()
regressor.fit(x_poly, y)
print('Accuracy:', regressor.score(x_poly, y))
```

```python
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x, y, color='g')
plt.plot(x, regressor.predict(x_poly), color='r')
```

```python
acc = []
for i in range(1, 21):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)
    regressor = LinearRegression()
    regressor.fit(x_poly, y)
    acc.append(regressor.score(x_poly, y))
acc
```

```python
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.ylim(min(acc), 0.984)
plt.bar(range(1, len(acc) + 1), acc, color='g')
```

```python
poly = PolynomialFeatures(degree=11)
x_poly = poly.fit_transform(x)
```
