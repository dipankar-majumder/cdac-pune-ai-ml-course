##### Image Cropping

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg')
img.shape
```

```python
new_img = img[200:450, 300:600]
cv2.imshow('Diagonal', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_15per = img[img.shape[0] * 15 // 100: img.shape[0] * (100 - 15) // 100, img.shape[1] * 15 // 100: img.shape[1] * (100 - 15) // 100]
print(img_15per.shape)
cv2.imshow('Diagonal', img_15per)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
1075 * (1 - 0.3)
```

```python
width, height, channels = img.shape
```

```python
startRow = int(width * 0.15)
startCol = int(height * 0.15)
endRow = int(width * 0.85)
endCol = int(height * 0.85)
```

```python
croppedImage = img[startRow:endRow, startCol:endCol]
cv2.imshow('Cropped Image', croppedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
patch = np.zeros((100, 100, 3), dtype=np.uint8) + (0x00, 0xFF, 0x00)
img[100:200, 100:200] = patch
cv2.imshow('Cropped Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
