##### 7. Apply the data encoding (if necessary)

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
x['SHAPE'] = label_encoder.fit_transform(x['SHAPE'])
x['SOIL_TYPE'] = label_encoder.fit_transform(x['SOIL_TYPE'])
```

```python
y = label_encoder.fit_transform(y)
y
```

##### 8. Implement the feature selection method chi2 and find top 5 features.

```python
from sklearn.feature_selection import SelectKBest, chi2
skf = SelectKBest(score_func=chi2, k=5)
skf.fit(x, y)
x_new = x.loc[:, skf.get_support()]
x_new
```

##### 9. Apply any feature scaling technique and transform

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_new)
x_scaled
```

##### 10. Perform the cross validation with split ratio 80:20 and stratification

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

##### 11. Implement all classification algorithm to check the accuracy on the data.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
nb = GaussianNB()
log = LogisticRegression(penalty='l2', solver='newton-cg')
knn = KNeighborsClassifier(n_neighbors=5)
dt = DecisionTreeClassifier()
svm = SVC(kernel='linear')
rf = RandomForestClassifier()
```

```python
nb.fit(x_train, y_train)
log.fit(x_train, y_train)
knn.fit(x_train, y_train)
svm.fit(x_train, y_train)
dt.fit(x_train, y_train)
rf.fit(x_train, y_train)
```

```python
y_pred_nb = nb.predict(x_test)
y_pred_log = log.predict(x_test)
y_pred_knn = knn.predict(x_test)
y_pred_svm = svm.predict(x_test)
y_pred_dt = dt.predict(x_test)
y_pred_rf = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Gaussian Naive Bayes:', accuracy_score(y_test, y_pred_nb))
print('Logistic Regression:', accuracy_score(y_test, y_pred_log))
print('K Nearest Neighbor:', accuracy_score(y_test, y_pred_knn))
print('Support Vector Machine:', accuracy_score(y_test, y_pred_svm))
print('Decision Tree:', accuracy_score(y_test, y_pred_dt))
print('Random Forest:', accuracy_score(y_test, y_pred_rf))
```

##### 12. Create the dataframe which shows the comparison of accuracy of different classification algorithm implemented on this dataset.

```python
accuracy_df = pd.DataFrame(data = [['Gaussian Naive Bayes', accuracy_score(y_test, y_pred_nb)],
                     ['Logistic Regression', accuracy_score(y_test, y_pred_log)],
                     ['K Nearest Neighbor', accuracy_score(y_test, y_pred_knn)],
                     ['Support Vector Machine', accuracy_score(y_test, y_pred_svm)],
                     ['Decision Tree', accuracy_score(y_test, y_pred_dt)],
                     ['Random Forest', accuracy_score(y_test, y_pred_rf)]], columns=['Algorithms', 'Accuracy'])
accuracy_df
```
